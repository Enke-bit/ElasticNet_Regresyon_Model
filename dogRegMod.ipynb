```python
import numpy as np
import pandas as pd
# Burda elestik net fonksyonunu.
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score
import matplotlib.pyplot as plt
# Elastiknet cv fonksyonunu import etme işlemlerini yapmaya çalışıyorum.
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV
```

```python
# Veri setinin okunmasını sağlıyorum.
df = pd.read_csv(r"C:\Users\realb\Desktop\ML_dersler\ML_201\ElasticNet Regresyon Model\Hitters.csv")

# Veri setinin içinden eksik gözlemleri çıkarmak gerekiyor bu işlemi yapıyorum.
df = df.dropna()

# Şimdi ise dummies ile kukla veri işlemi yapıyorum. One end coding yöntemi ile bu işlemi yapıyorum.
dms = pd.get_dummies(df[['League', 'Division','NewLeague']])

# Bağımlı değişkenimi "y" atıyorum.
y = df["Salary"]

# Veri setimiz içindeki bağımlı değişkeni ve katogorik değişkenlerin götürülmesi işlemini yapıyorum.
X_ = df.drop(['Salary', 'League', 'Division', 'NewLeague'], axis=1).astype('float64')

# Sonuçunda yukardaki değerlerin dummi versyonlarını ve bağımsız değişkenlerimi bir arada tutma işlemini yapıyorum.
X = pd.concat([X_, dms[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)

# Şimdi ise veri setimizde train ve test ayrımı yapmak için bu ayrımı yapma işlemini yapıyorum.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Şimdi model kurma işlemini gerçekleştiriyorum.
enet_model = ElasticNet().fit(X_train, y_train)
```

```python
# Model içinden kat sayılara bakıyorum.
enet_model.coef_
```

```python
# Sabit teriminede erişebilirim. 
enet_model.intercept_
```

```python
# ŞİMDİ KURDUĞUM BU MODEL İLE BİR TAHMİN ETME İŞLEMİ GERÇEKLEŞTİRİYORUM.
# Bu tahmin işlemini train üzerinden gerçekleştiriyorum.
enet_model.predict(X_train)[0:10]
```

```python
# Şimdi test veri setimiz üzerinden bu işlemi gerçekleştirebilirim.
enet_model.predict(X_test)[0:10]
```

```python
# Şimdi bu tahmin edilen değerler üzerinden hatamı hesaplamaya çalışıyorum.
y_pred = enet_model.predict(X_test)
# Bu alandan hata kareler ortalamasını hesaplamaya çalışyorum.
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
# R2 skorunu bu alandan hesaplamaya açalışıyorum.
r2_score(y_test, y_pred)
```
